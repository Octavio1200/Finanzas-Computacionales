## **4. Generar normal multivariada, generar precios sintéticos y valores del portafolio**

```python
# Simulación de Monte Carlo con la normal multivariada
# Semilla: resultados iguales
np.random.seed(12342)
#
mu = np.array([mediaRA1, mediaRA2])
# Número de simulaciones
S = int(1e5)
#  Rendimientos simulados
y = np.random.multivariate_normal(mu, SigmaA, S)
# Vector de precios
P = np.array([last_pA1, last_pA2])
# Posición inicial
PosIni = last_pA1 + last_pA2
# Posición en número de unidades de cada activo
x = np.array([10000, 10000])
# Portfolio al tiempo t=0
Port = np.dot(P, x)
# Precios simulados transpuestos
Psim = (P * np.exp(y)).T
# Precios simulados de los activos 1 y 2
PsimA1 = Psim[0]
PsimA2 = Psim[1]
# Valor de portafolio simulado
PortSim = np.dot(Psim.T, x)
```

```python
print(y.shape) # Rendimientos simulados
print(P.shape) # Dimensión vector de dos últimos valores
print(Psim.shape) # Dimensión de matriz de precios simulados
print(PsimA1.shape) # Primer vector columna de la matriz
print(PsimA2.shape) # Segundo vector columna de la matriz
print(PortSim.shape) # Dimensión de matriz de la posición del portafolio
```

```python
# Calcular CVaR del portafolio y de cada activo al 97.5%
#
alpha2 = 2.5
PLPort = np.column_stack((q, x[0] * (PsimA1 - last_pA1), x[1] * (PsimA2 - last_pA2)))
numPfPL = PLPort.shape[1]
VaRsMC = np.zeros(numPfPL)
CondVaRsMC = np.zeros(numPfPL)
for i in range(numPfPL):
    VaRsMC[i] = np.percentile(PLPort[:, i], alpha2)
    CondVaRsMC[i] = np.mean(PLPort[PLPort[:, i] <= VaRsMC[i], i])
```

```python
# Imprimir resultados adicionales
print('-------------------------------------------')
print('=== CVaR SIMULACIÓN MONTECARLO al 97.5% ===')
print('======= En unidades monetarias =======')
print('======= por Función percentile ==========')
print('-------------------------------------')
print(f'CVaR-Portafolio: {CondVaRsMC[0]:.4f}')
print(f'CVaR Activo 1: {CondVaRsMC[1]:.4f}')
print(f'CVaR Activo 2: {CondVaRsMC[2]:.4f}')
```
