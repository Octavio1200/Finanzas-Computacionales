# Introducción
El método de Box-Müller es un método de generación de pares de números aleatorios independientes con distribución normal
estándar, el método parte de una transformación de coordenadas rectangulares a polares.

El procedimiento general consiste en primer lugar al generar números aleatorios que se originen a partir
de la distribución uniforme, posteriormente se les aplica una transformación que los convierta en los números aleatorios de
una normal estándar. Recordemos que una variable aleatoria $X$ está distribuída normalmente con media media
$\mu$ y varianza $\sigma^2$, si su función de densidad de probabilidad está dada por:
$$
f(x)=\frac{1}{\sqrt[]{2 \pi \sigma}} e^{-\frac{1}{2}\left( \frac{(x-\mu)^{2}}{\sigma^{2}} \right)}
$$

Ahora, sean $X$ y $Y$ variables aleatorias normales unitarias independientes y sean $R$ y $\theta$
las coordenadas polares del vector $(X, Y)$. Es decir:

$$
R^{2} =X^{2} +Y^{2};
\tan \theta= \left( \frac{Y}{X} \right)
$$
Como $X$ y $Y$ son variables aleatorias independientes, su función de densidad conjunta es el producto de sus marginales, y
por lo tanto está dada por:
$$
f(x,y)=\frac{1}{\sqrt{2 \pi}} e^{-\frac{x^{2}}{2}}  \frac{1}{\sqrt{2 \pi}} e^{-\frac{y^{2}}{2}}
$$
Es decir:
$$
f(x,y)=\frac{1}{\sqrt{2 \pi}} e^{-\frac{(x+y)^{2}}{2}}
$$
Para determinar la densidad conjunta de $R^2$ y $\theta$, denotada por $f(d,\theta)$, se plantea el siguiente cambio de variables:

$$
d =x^{2} +y^{2}; \phantom{asd}\theta=\arctan (\frac{y}{x})
$$
Observamos que el cambio de variable está expresado en las dos funciones en términos de $x$ e $y$
para luego calcular el Jacobiano. Antes de encontrar el Jacobiano, vamos a calcular las  derivadas parciales de $d$ y $\theta$ con respecto de $x$ e $y$. Como $d =x^{2} +y^{2}$  
$$
\frac{\partial d}{\partial x}=2x ; \phantom{abc} \frac{\partial d}{\partial y}=2y
$$
De igual modo, como tenemos:
$$
\theta = \arctan (\frac{y}{x})
$$
Con ello:
$$
\frac{\partial \theta}{\partial x}= \frac{1}{1+\left( \frac{y}{x} \right)^{2}} \left( \frac{-y}{x^{2}} \right)
$$

$$
\frac{\partial \theta}{\partial y}= \frac{1}{1+\left( \frac{y}{x} \right)^{2}} \left( \frac{1}{x} \right)
$$
Luego de haber calculado las derivadas parciales procedemos a calcular el determinante de ellas, aplicando la definición anterior del Jacobiano pero con respecto a dos variables $x$ e $y$.
$$
J_{xy}^{f}=\frac{\partial (d, \theta)}{\partial (x, y)}=\left[ \begin{matrix}
2x &  \frac{1}{1+\left( \frac{y}{x} \right)^{2}} \left( \frac{-y}{x^{2}} \right)\\
2y & \frac{1}{1+\left( \frac{y}{x} \right)^{2}} \left( \frac{1}{x} \right)
\end{matrix} \right]
$$

$$
J_{xy}^{f} =2x \frac{1}{1+\left( \frac{y}{x} \right)^{2}} \left( \frac{1}{x} \right)  - 2y \frac{1}{1+\left( \frac{y}{x} \right)^{2}} \left( \frac{-y}{x^{2}} \right)= 2.
$$
El Jacobiano de esta transformación (es decir, el determinante de las derivadas parciales de $d$ y $\theta$ con  respecto de $x$, $y$ es igual a 2). Por esta ecuación se implica que la función de densidad conjunta de $R^{2}$ y $\theta$ está dada por:
$$
f(d, \theta)=\frac{1}{2} \frac{1}{2\pi}e^{-\frac{d}{2}};
0<d<∞, 0<\theta <2π
$$

Sin embargo, como esto es igual al producto de una densidad exponencial con media 2 (a saber, $\frac{1}{2} e^{-d/2}$) y la densidad uniforme en $(0, 2\pi)$ (a saber, $1/2\pi$), esto implica que: $R^{2}$ y $\theta$ son independientes; $R^{2}$ es exponencial con media 2 y $\theta$ se distribuye uniformemente en $(0,2\pi)$.

Como las variables $R$ y $\theta$ son independientes, sus funciones de densidad vienen dadas por:

$f(r)= r \times\exp \left\{ - \frac{r^{2}}{2} \right\} \Rightarrow  F(r)=1- \exp\left\{ - \frac{r^{2}}{2} \right\}$

$\Rightarrow R= \sqrt{-2 \ln(U)}$

$f(\theta)=\frac{1}{2 \pi} \Rightarrow F(\theta)=\frac{\theta}{2 \pi} \Rightarrow \theta = 2 \pi U$


Ahora podemos generar un par de variables aleatorias normales estándar independientes $X$ e $Y$, utilizando $R^{2}$ y
$\theta$ para generar primero sus coordenadas polares $\left( {X = R \cos \theta ;Y = R \sin \theta } \right)$ y luego
transformarlas de nuevo en coordenadas rectangulares.

De esta manera para generar variables aleatorias normales se aplica el siguiente algoritmo:
                                     
1.- Generar números aleatorios $U_1$ y $U_2$.

2.- ${R^2} = - 2\ln {U_1} \Rightarrow {R^2}$ es exponencial con media 2, $\theta  = 2\pi {U_2} \Rightarrow \theta $  es
uniforme en el intervalo  $(0,2\pi )$.

3.- Hacer:
$$
\begin{align*}
X &= R \cos \theta  = \sqrt { - 2\ln {U_1}} \cos \left( {2\pi {U_2}} \right)\\
Y &= R \;\mathrm{sen} \theta  = \sqrt { - 2\ln {U_1}} \sin \left( {2\pi {U_2}} \right)
\end{align*}
$$

## 1.Se cargan las paqueterias para trabajar con la generación de variables

```python
##
import numpy as np
from math import pi
from numpy import sqrt, sin, cos
from scipy.stats import norm
import scipy.stats as stats
import matplotlib.pyplot as plt
import matplotlib.colors as colors
```

## 2. Se ejecutan las transformaciones

```python
##
# Se generan las variables aleatorias.
n=2000
# Se generan las n muestras de U1 y U2, con U(0,1)
u1s = stats.uniform.rvs(size=n)
u2s = stats.uniform.rvs(size=n)

# Se transforma U1 a S...
ss=-np.log(u1s)

# Se transforma U2 a Theta...
thetas=2*pi*u2s

# Se convierte s a r...
rs=np.sqrt(2*ss)

# Se calcula a x e y a partir de theta...
X=rs*np.cos(thetas)
Y=rs*np.sin(thetas)

n_bins=100
# Para intervalos con regla de Sturges
# n_bins = int(np.ceil(np.log2(len(u1s))) + 1)
```

## 3. Se grafican los valores antes y después de la transformación.

```python
# Graficar las variables en una matriz de 2 x 2
# Gráfica de theta
plt.figure(figsize=(10,10))
plt.subplot(221) # Primer renglón
plt.hist(thetas, bins=n_bins, density=True, color="limegreen", edgecolor='white')
#plt.title(" $ \theta=2\pi U_2 $ ",fontsize=18)
plt.title(r'$\theta=2\pi U_2 $',fontsize=18)

# # Gráfica de R
plt.subplot(222)
plt.hist(rs, density=True,bins=n_bins, color="limegreen", edgecolor='white')
plt.title(r'$R= \sqrt {-2\ln U_1} $',fontsize=18)
overlayR=np.linspace(rs.min(), rs.max(), 100)
meanR, stdR=norm.fit(rs)
pdfR=norm.pdf(overlayR, meanR, stdR)
plt.plot(overlayR, pdfR, "r-", linewidth=3)

# Gráfica de X
plt.subplot(223)
plt.hist(X, bins=n_bins, density=True, color="limegreen", edgecolor='white')
plt.title(r'$X= R\cos \theta = \sqrt{-2\ln {U_1}} \cos({2\pi {U_2}})$',fontsize=18)
plt.xlim(-4,4)
overlay=np.linspace(X.min(), X.max(), 100)
mean, std=norm.fit(X)
pdf=norm.pdf(overlay, mean, std)
plt.plot(overlay, pdf, "r-", linewidth=3)

# Gráfica de Y
plt.subplot(224)
plt.hist(Y, bins=n_bins, density=True, color="limegreen", edgecolor='white')
plt.title(r'$Y= R\mathrm{sin} \theta = \sqrt{-2\ln {U_1}} \sin({2\pi {U_2}})$',fontsize=18)
plt.xlim(-4,4)
overlay=np.linspace(Y.min(), Y.max(), 100)
mean, std=norm.fit(Y)
pdf=norm.pdf(overlay, mean, std)
plt.plot(overlay, pdf, "r-", linewidth=3)
#plt.savefig('BM1jupyter.pdf',orientation='landscape')
#plt.savefig("BM1jupyter2.png")
#plt.savefig('BM1jupyter.eps',orientation='landscape', papertype='letter')
plt.show()
```

```python
# Esta parte estima los parametros de ss1
ss1=ss
param_Exp = stats.expon.fit(ss1)
print(param_Exp)
```

```python
##
# Prueba de normalidad
import pylab
jarque_bera_test_ss1,  p_value = stats.jarque_bera(ss1)
print('jarque_bera_test_ss1=%.4f, p=%.6f' % (jarque_bera_test_ss1, p_value))
if p_value > 0.05:
    print('Probablemente Normal.')
else:
    print('Probablemente no Normal.')
```

```python
# Ajustamos una exponencial
lexp1 = np.linspace(ss1.min(), ss1.max(), 100)
pdf_lexp_fitted1 = stats.expon.pdf(lexp1, *param_Exp)
```

```python
# Divide en 100 partes iguales a intervalo min y max de ss1
# Evalúa en la función de densidad de la exponencial
# Hace histograma
plt.figure(num=1, figsize=(10, 7))
plt.plot(lexp1, pdf_lexp_fitted1, color='r', lw=6, label="Exponencial")
plt.hist(ss1, density=True, bins=100, facecolor="limegreen", edgecolor='w',label="Frecuencia de ss1")
plt.legend(fontsize=18);
plt.title("Histograma vs Ajuste con Exponencial");
plt.xlabel("exp")
plt.show()
```

```python
# Ajustamos lognormal a muestras generadas de rs, hacer histograma y función de densidad.
# lognorm.fit realiza el ajuste de máxima verosimilitud, optimizando
# los parámetros para que la distribución lognormal se ajuste a los datos.
param_LN = stats.lognorm.fit(rs, floc=0.)  # floc=0 fija "loc" en 0 (opcional))
print(param_LN);
print("Parámetros estimados: ", param_LN)
print("Forma: ", param_LN[0])
print("Sigma: ", param_LN[1])
print("Escala: ", param_LN[2])
#
l = np.linspace(rs.min(), rs.max(), 100)
pdf_LN_fitted = stats.lognorm.pdf(l, *param_LN)
#
plt.figure(num=1, figsize=(10, 7))
plt.plot(l, pdf_LN_fitted, color='r', lw=3, label="Log-Normal")
plt.hist(rs, density=True, bins=100, facecolor="limegreen", edgecolor='w',label="Frecuencia de rs")
plt.legend(fontsize=18);
plt.title("Histograma vs Log-Normal");
plt.xlabel("rs")
plt.show()
```
