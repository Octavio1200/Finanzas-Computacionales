El [Modelo de Vasicek](https://en.wikipedia.org/wiki/Vasicek_model)
se describe mediante la ecuación diferencial estocástica:

$$
dr_t = a(b - r_t) dt + \sigma dW_t.
$$

Los parametros son:
- $a > 0$:  velocidad de reversion (speed of reversion).
- $b \in \mathbb{R}$:  nivel de largo plazo (long term mean level).  
- $\sigma > 0$:  volatilidad.

Es un proceso Gaussiano, Markoviano y estacionario (incondicional).

El modelo de Vasicek puede plantearse en términos discretos como una ecuación estocástica en diferencias:
$$
r_t=ab+(1-a)r_{t-1}+\sigma\epsilon_t
$$

donde $\epsilon_t\sim N(0,1)$. Al denotar $\beta_0=ab$  y $\beta_1=1-a$, entonces:
$$
r_t=\beta_0 + \beta_1r_{t-1}+\varepsilon_t
$$

donde ${\varepsilon_t}$  son variables aleatorias independientes y normalmente distribuídas con media cero
y varianza $\sigma^2$, $\varepsilon_t\sim N(0,\sigma^2)$.

Note que
$$
\widehat{a}=1-\widehat{\beta_1}
$$

y
$$
\widehat{b}=\frac{\widehat{\beta_0}}{a}.
$$


## **1. Resumen de la teoría sobre modelo de Vasicek**

- El modelo de Vasicek forma parte de los llamados modelos de equilibrio general debido al uso de condiciones de no arbitraje
para caracterizar el precio de un bono cupón cero a un plazo dado.

- Bajo los supuestos de equilibrio general y tasa corta neutral al riesgo, se resuelve la ecuación diferencial parcial parabólica del comportamiento del precio de un bono cupón cero.

- El precio de un bono cupón cero que se fija en $t$ y que al vencimiento $T$ paga una unidad monetaria, $B(T,T)=1$, se
denota por:$B=B(r_t,t;T)$ o en forma más simple como $B=B(t,T)$.

- En el caso del modelo de Vasicek el precio $B$ de un bono  cupón cero satisface:    

This is equation (1).
$$
{{{\partial B}\over {\partial t}}+\hbox{$1 \over 2$}\sigma^{2}{{\partial^2B} \over {\partial r^2}}+a(b-r_t){{\partial B}\over {\partial r_t}}-{r_tB}}=0.
$$

- La condición final corresponde al pago en el vencimiento del bono, $B(t,T)=1$. Las condiciones de frontera dependen de $a$, $b$, $\sigma$ y, por supuesto de, $r_t$.

- Dado que la ecuación  no cuenta con derivadas parciales cruzadas, se supone una solución en variables separables de  la siguiente forma:
    
$$
B(t,T)=e^{A(t,T)-r_tD(t,T)}.
$$

- Observe que en la fecha de vencimiento, necesariamente, $A(T,T)=0$ y $D(T,T)=0$, ya que $B(T,T)=1$.

- Las derivadas parciales de $B$ con respecto de $t$ y $r_t$, así como la segunda derivada parcial con respecto de $r_t$ están dadas por:
    
$$
\eqalign{&{{\partial B}\over {\partial t}}=\left({{\partial A}\over {\partial t}}-r_t{{\partial D}\over {\partial t}}\right)B,\cr &{{\partial B}\over {\partial r_t}}=-DB,\cr&{{\partial^2B} \over {\partial r_t^2}}=D^2B.}
$$
- Después de sustituir las ecuaciones anteriores en la EDP, se tiene que:
    
$$
{{{\partial A}\over {\partial t}}+ \hbox{$1 \over 2$}\sigma^{2}D^2-abD+r_t\left(-{{{\partial D}\over {\partial t}}}+aD-1\right)}=0.
$$    

- Si se deriva con respecto a $r_t$ se obtiene
$$-{{\partial D}\over {\partial t}}+aD-1=0$$
ó
$${{\partial D}\over {\partial t}}=aD-1.$$

- De hecho, la ecuación diferencial anterior es ordinaria y el uso de derivadas parciales es un abuso de notación.
La solución de la ecuación diferencial anterior con condición final $D(T,T)=0$ está dada por:
    
$$
\eqalign{D(t,T)=&D(T,T)e^{-a(T-t)}-e^{-a(T-t)}\int_T^te^{a(T-s)}d s\cr
=&-e^{-a(T-t)}\int_T^te^{a(T-s)}d s\cr
=&{1-e^{-a(T-t)}\over a}.}
$$   
y con base en la solución obtenida de $D(t,T)$ se obtiene:
  
$$
A(t,T) = \frac{1}{{{a^2}}}(D(t,T) - T + t)\left( {{a^2}b - \frac{1}{2}{\sigma ^2}} \right) - \frac{{{\sigma ^2}}}{{4a}}{D^2}(t,T).
$$

- **Curva de rendimiento**

A partir de los precios del bono para diferentes vencimientos $B(t,T), T\ge t$, se genera la estructura de plazos
de la tasa de interés $R(t,T)$ mediante la relación:
$$    
R(t,T) = - \frac{1}{{T - t}}\ln B(t,T).
$$
   
- Continuando con el desarrollo:
    
$$
R(t,T) = \frac{{{r_0} \times D(t,T) - A(t,T)}}{{T - t}}
$$
La tasa de interés de plazo más largo disponible en el mercado, también llamada tasa larga, $R(t,\infty)$, satisface
$$
R(t,\infty)= b  -{\sigma^2 \over 2a^2}.
$$
El precio de un bono cupón cero en términos de la tasa larga $R(t,\infty)$, con $B(t,T)=e^{-R(t,T)(T-t)}$, está dada por:

$$
B(t,T) = \exp \left\{ { - R(t,\infty )(T - t) - \left[ {{r_t} - R(t,\infty )} \right]D(t,T) - \frac{{{\sigma ^2}}}{{4a}}{D^2}(t,T)} \right\}.
$$
